```python
import numpy as np
import matplotlib.pyplot as plt

# Define the concentrations of the standard solutions
standards = [0, 5, 10, 20, 40, 60]

# Define the measured signals for each standard solution
signals = [0.2, 4.8, 9.3, 19.2, 38.1, 56.8]

# Calculate the slope and intercept of the linear regression line
slope, intercept = np.polyfit(standards, signals, 1)

# Generate the plot
plt.scatter(standards, signals, label='Data points')
plt.plot(standards, intercept + slope * np.array(standards), label='Linear regression')
plt.xlabel('Concentration (mg/L)')
plt.ylabel('Signal (mV)')
plt.title('Standard addition plot')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the concentrations of the standard solutions
standards = [0, 5, 10, 20, 40, 60]

# Define the true concentration of the sample
sample_true_conc = 30

# Define the signal-to-concentration ratio for the standards
ratio = 2

# Define the signal noise level
noise_level = 1

# Generate the measured signals for each standard solution
signals = [c * ratio + np.random.normal(scale=noise_level) for c in standards]

# Generate the measured signal for the sample, adding some random noise
sample_signal = sample_true_conc * ratio + np.random.normal(scale=noise_level)

# Generate the plot
plt.scatter(standards, signals, label='Standards')
plt.scatter(0, sample_signal, label='Sample')
plt.xlabel('Concentration (mg/L)')
plt.ylabel('Signal (mV)')
plt.title('Standard addition plot')
plt.legend()
plt.show()
```
