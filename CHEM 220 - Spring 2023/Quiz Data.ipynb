```python
import random
import numpy as np

test = []
length = 5

for x in range(length):
    
    sample = random.randrange(40000, 65000, 1)
    sample = sample/200000
    test.append(sample)

print(F"You determined the following concentrations by titration {test} M")
print(F"The average of the data set is {np.mean(test)}")
print(F"The standard deviation of the data set is {np.std(test)}")

```

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 12})

x = np.arange(0, 5.5, 0.5)




y = (x*10)+15

# Define the signal noise level
noise_level = 2

y = [(c * 10) + 15 + np.random.normal(scale=noise_level) for c in x]
y = [i/100 for i in y]

slope, intercept, r_value, p_value, m_err = stats.linregress(x,y)

print(m_err)

slope_str = str(round(slope,4))
int_str = str(round(intercept,3))
R_string = str(round(r_value**2,4))

string0 = "y = " + slope_str +" x + " + int_str
string2 = "R Squared =  "+ R_string

fig, ax = plt.subplots()
ax.scatter(x, y)

x_extended = np.arange(-5, 5.5, 0.5)

ax.plot(x_extended, slope*x_extended + intercept, color='red')


# Set the x and y axis ranges
ax.set_xlim([-2, 5.5])
ax.set_ylim([0, 0.9])

# Add axis labels and title
ax.set_xlabel('Concentration (mM)')
ax.set_ylabel('Absorbance')
ax.text(-1,0.7,string0)
ax.text(-1, 0.6,string2)
#ax.set_title('Sample Scatter Plot')

# Show the plot
plt.show()

```
