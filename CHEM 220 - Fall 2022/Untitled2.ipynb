## Playing around with the data frame to reorganize the sample data

```python
# This is a strategy for importing data into lists so that we can use it!
import numpy as np
import scipy.stats as stats
import pandas as pd
import math
# read the whole data file in:
data = pd.read_csv("Eucalyptus 0.1 ppm 9-6-22.csv", header = 6)


# read all elements and wavelengths in the file:
elements = data.Element.unique()
print(F'These are all the elements and wavelengths that appear in the data file {elements}')
```

```python
## slice by Sample name instead of other IDs

HCl = np.empty([len(elements)-1, 120])
#HNO3 = []
index = 0

for x in elements:
    
    if x != "Ar 430.010":
        temp = data[data.values == x]
        temps = temp[temp.values == 'Sample']
        labels = temps.Label.unique()
        #print(len(labels))
        #grab all intensity values
        #remove nonsense - what do you actually want me to do here?
        temps_Int = temps['Concentration'].replace(['-','####','Uncal'], 0)
        list_int = temps_Int.tolist()
        #make sure all values are actually numbers
        list_int = [float(j) for j in list_int]
        #print(len(list_int))
        for i in range(len(list_int)):
            if i < 30 and index < 34:
                #print(x,i)
                HCl[index,i] = list_int[i]
        index = index +1
            
                
        
#print(HCl)      
for x in range(len(elements)):
    print(F'The HCl treated samples for {elements[x]} are {HCl[:,x]}')
            
```
