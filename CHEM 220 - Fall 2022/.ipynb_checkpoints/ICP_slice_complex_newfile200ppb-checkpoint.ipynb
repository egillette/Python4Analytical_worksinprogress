## Part 1 - Reading and Organizing the data

This section reads in the csv file, sorts each line by element and whether or not it is a standard or a sample. Currently all other data types (QC etc are ignored). Argon is removed from the elements list and data lists in this block as well. Current output is just confirming all of the unique elements identified in the file, and giving the standard values for the first element, as a place to double check that the code is working.

```python
# This is a strategy for importing data into lists so that we can use it!
import numpy as np
import scipy.stats as stats
import pandas as pd
import math
# read the whole data file in:
data = pd.read_csv("Eucalytpus 0.2 ppm 12-15-22 Successful.csv", header = 6)


# read all elements and wavelengths in the file:
elements = data.Element.unique()
print(F'These are all the elements and wavelengths that appear in the data file {elements}')

standards =[]
conc = []

#slice by element - all standards data, exclusing argons, since it's always weird?

for x in elements:
    if x != "Ar 430.010":
        temp = data[data.values == x]
        temps = temp[temp.values == 'STD']
        #grab all intensity values
        #remove nonsense - what do you actually want me to do here?
        temps_Int = temps['Intensity'].replace(['-','####'], 0)
        #convert to list
        list_int = temps_Int.tolist()
        #make sure all values are actually numbers
        list_int = [float(j) for j in list_int]
        #add this list to the end of the existing list of standard intensities
        standards.append (list_int)
        
        #grab concentration values
        #remove nonsense - what do you actually want me to do here?
        temps_conc = temps['Concentration'].replace(['-','####','--'], 0)
        # convert to list
        list_conc = temps_conc.tolist()
        # make sure all list digits are numbers
        list_conc = [float(j) for j in list_conc]
        #add to the end of exisiting list of concentrations
        conc.append(list_conc)
    
#show one example to check values
print(F'These are all the standard intensities for {elements[0]} are {standards[0]}')
print (F'These are the concentrations for {elements[0]}, as recorded in the file {conc[0]}')


#repeat for samples instead of standards
sample = []
sample_label = []


#slice by element - all sample data
for x in elements:
    if x != "Ar 430.010":
        temp = data[data.values == x]
        temps = temp[temp.values == 'Sample']
        temps_clean = temps['Intensity'].replace(['-','####'], 0)
        
        temps_Int = temps_clean.tolist()
        
        #make sure all values are actually numbers
        temps_Int = [float(j) for j in temps_Int]
    
        sample.append (temps_Int)
        
        #repeat to get sample labels
        temp = data[data.values == x]
        temps = temp[temp.values == 'Sample']
        temps_label = temps['Label'].tolist()
    
        sample_label.append (temps_label)

        
#why is Zn 216 not working? ALL LABELS ARE OFF BY ONE, BECAUSE OF ARGON :-(

#quick and dirty solution - remove argon from the list

elements = np.delete(elements, 3)

#print(F'All of the samples for {elements[0]} are {sample_label[0]} with values of {sample[0]}')



    

```

## Standard Curve Fitting
All standards data are processed here, linear regressions applied, and best-fit lines added to graphs, with equations of the line output above the relevent standard curve.

```python
import matplotlib.pyplot as plt



x = np.array(conc, dtype = object)
y = np.array(standards, dtype = object)

#empty space for unknown concentrations

unknowns = []


for i in range(len(elements)):
    m, b, R2, p, s_m = stats.linregress(x[i], y[i])
    
    # solve for unknown_concentration
    unknown_y = np.array(sample[i])
    unknown_concentration = (unknown_y-b)/m
    unknowns.append(unknown_concentration)

    print (F'the equation of the line for {elements[i]} is y = {m}x + {b} and its R-squared value is {R2}')
    # Use our predicted model
    x_predict = np.array(x[i])
    predict_y = (x_predict*m)+b
    # Plot the predicted data (note if we don't ask for a specific marker style, we'll just get a line)
    plt.plot(x[i],predict_y)
    plt.plot(x[i], y[i], 'rx')
    
    plt.xlabel(F'concentration of {elements[i]}')
    plt.ylabel(F'intensity of {elements[i]}')
    plt.show()
    

    

#plt.errorbar(x.T, y.T, yerr = None, xerr = None, fmt = 'r.')
#plt.xlabel("concentration (ug/L)")
#plt.ylabel("average intensity")
```

## Part 4 - Organizing samples by element for all treatments

This block has the graphs of each of the 4 treatments for each element, with a legend. Currently, the x and y axis ranges are locked, so if a graph appears empty, it is because the data does not fall in the y range 0 to 0.11. This can be updated, as desired.

## playing around with scip polyfit

Curve fitting in enabled here; the data remains organized as above (all treatments graphed together for each element). Best fit lines based on the current curve fitting function are added to the graphs, and the variable from the fit are printed above the graph. Note treatments are in the same order as above, so treatement 0 is HCl, 1 is HNO3, 2 is NaOH and 3 is Control. 

```python
from scipy.optimize import curve_fit
#this is the test function - the variable has to go first followed by as many parameter variables as you will need. Here we use 3, called a c and d (skipping b since it's used above in the linear regressions)
def testfunction(ex,a,c,d):
    return a*np.exp(-c*ex)+d

#data organization as before    
for x in range(len(elements)):
    #print(F'the concentration of {elements[x]} in these samples')
    label = sample_label[x]
    number = unknowns[x]
    n = number.reshape(-1,3)
         
    sample = n.reshape(-1, 3).mean(axis=1)
    all_sample = sample.reshape(-1,10)
     
    
    
    # for curvefit to work, we need the ex variable as a numpy array
    #time is the actual data x axis
    time = np.array([0,15,30,45,60,90,120,240,480,720])
    
    #xx is a list of numbers from 1 to 720 which is used to draw detailed best fit lines
    xx = np.array(list(range(1,720)))
    # an index of colors is used here to make sure markers for the actual data and best fit lines always match in color!
    colors = ['r','k','b','g'] # red, black, blue then green)
    
    for i in range(len(all_sample)):
        # again, just making things np arrays
        yfit = np.array(all_sample[i])
        #here we apply the curve fit. The format is the model function, the x dat and y data we're trying to fit, the p0 = starting guesses for variables, maxfev= # of iterations allowed before the regression fails
        params, cv = curve_fit(testfunction, time, yfit, p0 = [0.5,0.0025,0.01],maxfev=8000)
        #here we take the best fit parameters the regression determines, and print them out:
        print(F'The fittings paratmers a, c and d are {params} for {elements[x]} treatment number {i}')
        
       #plug the parameters back into the test function with the dummy x data to draw detailed best fit lines
        predict_y = testfunction(xx,*params)
        #plot the real data, with circles
        plt.plot(time, yfit, 'o', color=colors[i%4])
        #plot the best fit line, as a line
        plt.plot(xx,predict_y, color=colors[i%4])
        #axis labes and legends
        plt.xlabel(F'time')
        plt.ylabel(F'concentration of {elements[x]}')
        plt.legend(['HCl', 'HCl fit', 'HNO3','HNO3 fit', 'NaOH', 'NaOH fit', 'Control', 'Control fit'])
    plt.show()           
            
        
        
        
```
